```python
import urllib.request
import shutil
import os
```

```python
# create a new directory within Catalog
directory = "/Volumes/ny_taxi/data/source_files/NYC_Taxi_yellow_2025"
os.makedirs(directory, exist_ok=True)
```

```python


for i in range(1,7):
    url = f"https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2025-0{i}.parquet"
    with urllib.request.urlopen(url) as u,open(directory+f"/yellow_tripdata_2025-0{i}.parquet", "wb") as f:
        shutil.copyfileobj(u,f)
    
```
