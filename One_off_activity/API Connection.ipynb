```python
import urllib.request
import shutil
import os
```

```python
# create a new directory within Catalog for Yellow-Taxi-data
directory = "/Volumes/ny_taxi/data/source_files/NYC_Taxi_yellow_2025"
os.makedirs(directory, exist_ok=True)
```

```python
# loop through URL, open each URL and write binary data into volume

for i in range(1,7):
    url = f"https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2025-0{i}.parquet"
    with urllib.request.urlopen(url) as r,open(directory+f"/yellow_tripdata_2025-0{i}.parquet", "wb") as f:
        shutil.copyfileobj(r,f)
    
```

```python
# create a new directory within Catalog fpr Look-up table
directory_2 = "/Volumes/ny_taxi/data/source_files/NYC_Taxi_lookUp_tables"
os.makedirs(directory_2, exist_ok=True)
```

```python
url = 'https://d37ci6vzurychx.cloudfront.net/misc/taxi_zone_lookup.csv'
with urllib.request.urlopen(url) as r,open(directory_2+'/look_up.csv', "wb") as f:
        shutil.copyfileobj(r,f)
```
